```python
import pandas as pd

url = "https://raw.githubusercontent.com/mevangelista-alvarado/datasets/refs/heads/main/spotify_songs.csv"
df = pd.read_csv(url)
#importo pandas y cargo un archivo CSV desde una URL que contiene información de canciones de Spotify
```

```python
df #para ver los datos
```

```python
# Seleccionar características (features)
features = [
    'danceability', 'energy', 'key', 'loudness',
    'mode', 'speechiness', 'acousticness', 'instrumentalness',
    'liveness', 'valence', 'tempo', 'duration_ms',
]
X = df[features].values #elijo las columnas que voy a usar como características (por ejemplo, energía, tempo, duración, etc.) y guardo sus valores en x

# Target numérico
y = df['popularity'].values   # La columna "popularity" es lo que quiero predecir, así que la guardo en y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(  #Divido los datos en dos grupos
    X, y,
    test_size=0.2, random_state=42
)
# Uso random_state=42 para que la división sea siempre igual
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  #La primera tiene 64 neuronas
    Dense(32, activation='relu'),   #La segunda tiene 32 neuronas
    Dense(1, activation='linear')   # La última tiene 1 neurona
]) #creo una red neuronal con tres capas
```

```python
from tensorflow.keras.optimizers import Adam

# Tasa de aprendizaje deseada
learning_rate = 0.001
adam_optimizer = Adam(learning_rate=learning_rate)  #Uso el optimizador Adam con una tasa de aprendizaje pequeña (0.001), que ayuda a ajustar los pesos del modelo poco a poco
```

```python
model.compile(
    optimizer=adam_optimizer,
    loss='mse',
    metrics=['mae'],
) #Compilo el modelo
```

```python
history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=50,
    batch_size=50,
)  # Entreno el modelo durante 50 épocas
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Pérdida de entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.legend()
plt.title('Función de pérdida durante el entrenamiento')
plt.show()
#Genero una gráfica para ver cómo fue bajando el error durante el entrenamiento y la validación
```

```python
loss, mae = model.evaluate(X_test, y_test)
print(f"MAE en el conjunto test: {mae}") #Evalúo el modelo con los datos de prueba y muestro el MAE, que indica qué tan cerca están las predicciones de los valores reales
```

```python
import pandas as pd #Uso el modelo para predecir la popularidad

predictions = model.predict(X_test)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': predictions.flatten()})
print(comparison.head())
```

```python
# indice a selecionar
i = 0
cancion = canciones_df.iloc[i]
X_input = cancion[features].values.reshape(1, -1)
X_input = scaler.transform(X_input)

prediccion = model.predict(X_input)[0][0]
print(f"Canción: {cancion['track_name']} - {cancion['artists']}")
print(f"Popularidad real: {cancion['popularity']}")
print(f"Predicción: {prediccion:.2f}")
#pido al modelo que prediga su popularidad
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2 = r2_score(y_test, predictions)
print(f'R²: {r2}')

mse = mean_squared_error(y_test, predictions)
print(f'MSE: {mse}') #
```
